```python
import pickle
import pandas as pd
```

```python
finGpt_sentiment = {}
with open(f"../res/evo.pkl","rb") as f:
  finGpt_sentiment = pickle.load(f)

davinci_sentiment = {}
with open(f"../../data/text-davinci-003.pkl","rb") as f:
  davinci_sentiment = pickle.load(f)

curie_sentiment = {}
with open(f"../../data/text-curie-001.pkl","rb") as f:
  curie_sentiment = pickle.load(f)

fingpt_df = pd.DataFrame(list(finGpt_sentiment.items()), columns=['News', 'Sentiment'])
davinci_df = pd.DataFrame(list(davinci_sentiment.items()), columns=['News', 'Sentiment'])
curie_df = pd.DataFrame(list(curie_sentiment.items()), columns=['News', 'Sentiment_curie'])

sentiment_df = fingpt_df.merge(davinci_df, on='News', how='inner', suffixes=("_fingpt", "_davinci")) 
sentiment_df = sentiment_df.merge(curie_df,on='News', how='inner')

sentiment_df
```

```python
def map_sentiment(sentiment):
    sentiment_lower = sentiment.lower()
    if 'positive' in sentiment_lower:
        return 1
    elif 'neutral' in sentiment_lower:
        return 0
    elif 'negative' in sentiment_lower:
        return -1
    else:
        return sentiment 

sentiment_df['Sentiment_fingpt_number'] = sentiment_df['Sentiment_fingpt'].apply(map_sentiment)
sentiment_df['Sentiment_davinci_number'] = sentiment_df['Sentiment_davinci'].apply(map_sentiment)
sentiment_df['Sentiment_curie_number'] = sentiment_df['Sentiment_curie'].apply(map_sentiment)
sentiment_df['missing'] = sentiment_df['Sentiment_fingpt_number'].apply(lambda x: not isinstance(x, (int, float)))
sentiment_df[sentiment_df['missing'] == False]
```

```python
missing_df = sentiment_df[sentiment_df['missing'] == True]
missing_df
```

```python
columns = ['id', 'News', 'Date', 'Time', 'User_id']
df_original = pd.read_csv('../df_original.csv', names=columns, header=0)

missing_df = missing_df.merge(df_original, on='News', how='inner')
```

```python
missing_sentiment = {}
with open(f"../res/missing.pkl","rb") as f:
  missing_sentiment = pickle.load(f)

missing_df = pd.DataFrame(list(missing_sentiment.items()), columns=['News', 'Sentiment'])
missing_df.iloc[5]['Sentiment']
```

```python
import re


def find_sentiment(sentiment):
  pattern = r'Sentiment:\s+(Positive|Neutral|Negative)'
  match = re.search(pattern, sentiment)

  if match:
      # sentiment
      return match.group(1)
  return "NotFound"

missing_df['ps'] = missing_df['Sentiment'].apply(find_sentiment)
missing_df['ps'].str.contains('NotFound').value_counts()
```
